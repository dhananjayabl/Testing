```python
import torch
from transformers import AdamW, AutoTokenizer, AutoModelForSequenceClassification
```

```python
chk_pt = 'michiyasunaga/BioLinkBERT-large'
tokenizer = AutoTokenizer.from_pretrained(chk_pt)
model = AutoModelForSequenceClassification.from_pretrained(chk_pt)
```

```python
sequences = ["PROTOCOL: Med Unit V: Amivantamab (week 4 and subsequently 2 weekly)",
"To start on (Cycle 25) - 15-07-2023-in Special Daycare",
"Premedications (prior to amivantamab):",
"o Inj Dexa 10 mg IV (60 min prior to Amivantamab)",
"o Inj Avil 25mg IV (30 min prior to Amivantamab)",
"o Inj paracetamol 1000mg IV (30 min prior to Amivantamab)",
"o Inj Emset (Ondansetron) 8 mg IV (15 min prior to Amivantamab)",
"o Inj Rantac 50 mg in 100ml NS (15 min prior to Amivantamab)",
"Amivantamab Infusion (can be given through a central line or peripheral line): Doctor to be present and check emergency tray prior to starting",
"o Inj Amivantamab-1050 mg-(if < 80Kg) in 250 ml NS (PVC bag) (make upto 250ml final volume) through codon set (please prime with NS 1st) IV at 125 ml/hour for 2 hours",
"o 100ml NS IV flush at 125 ml/hour (same rate as drug ongoing infusion)]"]
```

```python
labels = [18, 18, 15, 15, 15, 15, 15, 15, 18, 18, 18]
print(len(labels))
```

```python
batch = tokenizer(sequences, padding=True, max_length=512, truncation=True, return_tensors='pt')
```

```python
# prompt: create a label mapping code from labels

label_to_id = {label: i+1 for i, label in enumerate(labels)}
label_to_id
```

```python
optimizer = AdamW(model.parameters(), lr=5e-5)
loss  = model(**batch).loss
loss.backward()
optimizer.step()
```
