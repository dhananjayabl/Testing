```python
import pandas as pd
import random
from datetime import datetime, timedelta
import numpy as np

# Defining sample data for the columns
patient_ids = range(1, 101)  # 100 unique patient IDs
opd_ids = range(1, 21)  # 20 unique OPD IDs
doctor_ids = range(1, 21)  # 50 unique doctor IDs
investigation_types = ['Lab_Investigation', 'Imaging_Investigation']
investigation_names_lab = ['Blood Test', 'Urine Test', 'Liver Function Test', 'Kidney Function Test', 'Complete Blood Count']
investigation_names_imaging = ['X-Ray', 'CT Scan', 'MRI', 'Ultrasound', 'Mammogram']
# anatomical_parts_lab = ['Blood', 'Urine', 'Liver', 'Kidney']
anatomical_parts_imaging = ['Chest', 'Abdomen', 'Head', 'Pelvis', 'Breast', 'Lung', 'Mouth']
instructions = ['Fasting required', 'Drink water before test', 'Avoid eating for 12 hours', 'No special instructions', 'Do not wear any metal objects']

# Generating random sample data for investigation prescriptions
def generate_investigation_sample_data(num_samples):
    data = []
    for i in range(num_samples):
        patient_id = random.choice(patient_ids)
        date = datetime.now() - timedelta(days=random.randint(0, 365))
        opd_id = i
        doctor_id = random.choice(doctor_ids)
        investigation_type = random.choice(investigation_types)
        if investigation_type == 'Imaging_Investigation':
            investigation_name = random.choice(investigation_names_imaging)
            if investigation_name == 'Mammogram':
              anatomical_part = 'Breast'
            else:
              anatomical_part = random.choice(anatomical_parts_imaging)
        else:
            investigation_name = random.choice(investigation_names_lab)
            anatomical_part = np.nan
        instruction = random.choice(instructions)

        data.append([patient_id, date.strftime('%Y-%m-%d'), opd_id, doctor_id, investigation_type, investigation_name, anatomical_part, instruction])

    return data

# Generate 1000 sample rows
investigation_sample_data = generate_investigation_sample_data(1000)

# Create DataFrame
columns_investigation = ['Patient_Id', 'Date', 'OPD_Id', 'Doctor_Id', 'Investigation_Type', 'Investigation_Name', 'Anatomical_Part', 'Instructions']
df_investigation = pd.DataFrame(investigation_sample_data, columns=columns_investigation)
df_investigation.head()
```
