```
X_data = tf.random.normal([500, 1, 255])
Y_data = tf.random.uniform([500, 5])
```

```
Y_data[0]
```

```
def res_mode(input_shape, classes=5):
    k_init = keras.initializers.GlorotUniform(seed=0)
    
    model = Sequential()
    model.add(Conv1D(1024, kernel_size=1, input_shape=input_shape[1:], activation='relu'))
    model.add(Conv1D(1024, kernel_size=1, activation='relu'))
    model.add(Dropout(0.5))
    model.add(MaxPooling1D(pool_size=1))
    model.add(Flatten())
    model.add(Dense(1000, activation='relu'))
    model.add(Dense(classes, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    return model
```

```
model = res_mode(X_data.shape, classes=5)

history = model.fit(X_data, Y_data, epochs=20, batch_size=32)
```
